Take several pictures of red, blue, and green items with your phone or other digital camera (or download some from the internet, if a camera isn't available).

Load each image, and convert it to a tensor.

```python
import numpy as np
import torch
torch.set_printoptions(edgeitems=2, threshold=50)
```

```python
img_arr = {
    'red_item': imageio.imread('./red_item.jpg'),
    'green_item': imageio.imread('./green_item.jpg'),
    'blue_item': imageio.imread('./blue_item.jpg')
}
```

Converting to tensor

```python
for name, img in img_arr.items():
  img = torch.from_numpy(img).float()
  img = img.permute(2, 0, 1)    #(H, W, C) -> (C, H ,W)
  img_arr[name] = img
```

For each image tensor, use the `.mean()` method to get a sense of how bright the image is

```python
for name, img in img_arr.items():
  print(f'{name} mean: {torch.mean(img)}')
```

Take the mean of each channel of your images. Identify the red, green, and blue items from the channel averages.

```python
for name, img in img_arr.items():
  for channel in range(3):
    print(f'{name} mean: {torch.mean(img[channel])}')
  print()
```
